```python
maxF1Vec = mean(f1Cran + f1Cisi + f1Vas)
vecPrint = ('Vectorial', maxF1Vec)

maxF1LSA = mean(k50f1Cran + k100f1Cran + k150f1Cran + k200f1Cran +
                 k50f1Cisi + k100f1Cisi + k150f1Cisi + k200f1Cisi +
                 k50f1Vas + k100f1Vas + k150f1Vas + k200f1Vas)
lsaPrint = ('LSA', maxF1LSA)

maxF1Bool = mean([0.02, 0.02, 0.002, 0.04, 0, 0.02])
boolPrint = ('Booleano', maxF1Bool)

#result = [vecPrint, lsaPrint, boolPrint]

maxF = [vecPrint, lsaPrint, boolPrint]
maxF.sort(key=lambda x : x[1], reverse=True)
for i in maxF:
    print(f'Media de medidas F1 de {i[0]} es: {i[1]}')
```
